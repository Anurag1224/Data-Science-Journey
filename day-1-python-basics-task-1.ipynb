# 3. Variables

````python
# Static vs Dynamic Typing
# Static vs Dynamic Binding
# Stylish declaration techniques
````

````python
# Static Typing and Dynamic Typing


# In C/C++ (Here we need to define the data type int, char, float, bool etc. while declaring the variable)
# And this is known as ------Static Typing-----------

# int name = 'Anurag'

# In Python (we don't need to define the data type) - While interpreting the code the interpreter automatically understood it's type based on the value given to the variable
# And this is known as ------ Dynamic Typing-----------

name = 'Anurag'
print(name)

a = 5
b = 7
print(a+b)
````

````python
# Dynamic Binding

a = 5
print(a)

a = 'Anurag'
print(a)

# This is known as dynamic binding where the variable don't have fixed data type and can be override by any other value.

# Static Binding

# In programing languages like C/C++ or Java we have to define the data type of any variable and it's type is fixed for that scope and we can't override with any other data type value.
# int a = 10
# a = 20 ----------POSSIBLE
# but a = 'Anurag' -------------NOT POSSIBLE
````

````python
# Ways to declare variable

#normal way (4 lines)

a = 1
b = 2 
c = 3
print(a,b,c)

# Or compressed way (2 line)

a, b, c = 1, 2, 3
print(a,b,c)

# Another way if you want to assign same value to each variable
a = b = c = 5
print(a,b,c)
````

````python
# Comments (In python only we can write single line comment using hash and there is no multi line comment in  python)

# This is a comment
a = 4
b = 6 # This is a comment

# This is a second comment

print(a+b)
````

# 6. Type Conversion

````python
# Implicit vs Explicit

# Implicit type conversion - here we don't need to do type conversion. In implicit type conversion the interpreter do this automatically by itself
print(5+5.6)

print(type(5), type(5.6))
#here one is int and other is float and the end result is float. So here there is implicit type conversion where int is converted to float internally i.e implicitely.

#Note - Implicit type conversion does not work every time. Lets see here

print(4 + '4') # throws error

# to handle this we use explicit type conversion.
````

````python
# Explicit type conversion
# Let's convert string to integer

print(int('4'))
print(type(int('4')))

print(int(4.5))
print(type(int(4.5)))

#int to str
str(5)

#int to float
float(10)

# Similarly we can do for other data types also
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# lets assume point A(x,y) and point B(x1,y1). We have to find distance of AB
# Eucledian distance AB = sqrt((x2-x1)^2 + (y2-y1)^2)

import math

x = float(input("Enter the x cordinate of point A"))
y = float(input("Enter the y cordinate of point A"))
x1 = float(input("Enter the x cordinate of point B"))
y1 = float(input("Enter the y cordinate of point B"))

#AB = math.sqrt((x1-x)*(x1-x) + (y1-y)*(y1-y))
# we can use exponent ** instead of repeating multiplication for power i.e **2 for square
#we need to import the math module

#normal method
AB = math.sqrt((x1-x)**2 + (y1-y)**2)

#using dist function to calculate the distance between two coordinates.
AB2 = math.dist((x, y), (x1, y1))

print("Distance between point AB = ", AB)
print("Distance between point AB2 = ", AB2)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# nth term of AP is given by formula : firstTerm + (noOfTerms-1)* commonDifference

a = int(input("Enter the first term of an AP"))
b = int(input("Enter the second term of an AP"))
n = int(input("Enter the no. of terms in an AP"))

# Calculate common difference
d = b - a

# Calculate nth term
lastTerm = a + (n-1)*d

print("Last term or nth of an AP is ", lastTerm)
````
